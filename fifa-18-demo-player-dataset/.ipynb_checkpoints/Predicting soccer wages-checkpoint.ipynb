### Age Distribution.

### Salary Distribution.

```python
# Set The players' value worth and the players' potential into numeric data cause at the moment the data
#is in string format.
wages = np.array(df['Wage'])
salary = []
for c in wages: #iterate over the wages list
    c = c.replace('€','') #remove the Euro symbol, by replacing it with nothing
    if 'K' in c:    #For Thousands
        salary.append(c.replace('K', "000")) #replace the 'K' with actual with zeros
    else:
        salary.append(c)
# change wage dtype to integers
wages = [int(i) for i in salary] 
```

```python
plt.hist(wages, bins=60, edgecolor="w")
plt.show()
```

```python
print(max(wages))
print(min(wages))
```

It Seems that most of our players earn between 0 Euros and 565 000 Euros, but it is highly unlikey that a pearson earn zero wages, and there must be something special about a person earning €565000... let's explore these **Outliers**.

---------
We'll explore the players that earn €565K in wages, we'll explore how to deal with them.

```python
# Outlier retrieval
df1.head(2)
high = df1[df1['Wage']== "€565K"]
high # call the outlier DataFrame.
```

```python
dset = high.drop(columns=['Name','Club','Wage'])
def to_numeric(df):
    """" This function takes a numerical object populated DataFrame and changes the num populated
    columns from object to int64"""
    """"
    df - This is DataFrame that only made up of numbers but the dtype of the DataFrame is not int64.
    """
    df_jojo = pd.DataFrame()
    if type(df) == type(df_jojo):
        for column in df:
            df[column] = pd.to_numeric(df[column])
        return df
```

```python
dset = to_numeric(dset)
```
